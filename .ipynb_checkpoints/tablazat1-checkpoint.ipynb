Normális táblázat: $P(Z<z)$ értékeket tartalmazza:
<img src="./normal01.jpg"></img>

Direkt keresés:
$$
P(Z<0.1)=0.5398
$$
$$
P(Z>0.45)=1-P(Z<0.45)=1-0.6736=0.3264
$$
$$
P(Z<-1.33)=P(Z>1.33)=1-P(Z<1.33)=1-0.9082=0.0918
$$
$$
P(Z>-1.65)=P(Z<1.65)=0.9505
$$


Fordított keresés. Kvantilis. Melyik az a $z$, melyre 
$P(Z<z_p)=p$, ahol $p$ egy adott valószínűség.
$$
P(Z<z)=0.95 \ \ \implies \ \ z\approx 1.65
$$
$$
P(Z<z)=0.90 \ \ \implies \ \ z\approx 1.28
$$
$$
P(Z<z)=0.975 \ \ \implies \ \ z\approx 1.96
